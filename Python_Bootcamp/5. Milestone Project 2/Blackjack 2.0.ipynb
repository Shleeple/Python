```python
from random import shuffle
ranks = ("Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace")
suits = ("Spades", "Hearts", "Clubs", "Diamonds")
values = {"Two":2, "Three":3, "Four":4, "Five":5, "Six":6, "Seven":7, "Eight":8, 
          "Nine":9, "Ten":10, "Jack":10, "Queen":10, "King":10, "Ace":11}
```

```python
class Card():
    
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit
        self.value = values[rank]
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"
    
```

```python
class Deck():
    
    def __init__(self):
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                created_card = Card(rank, suit)
                self.all_cards.append(created_card)
    
    def deal_one(self):
        return self.all_cards.pop(0)
    
    def shuffle(self):
        shuffle(self.all_cards)
        
```

```python
# GAME LOGIC

game_on = True

while game_on:
    
    # Shows the player how many chips they have
    print(f"Player has {len(player.money)} chips")
    
    # Ask player for bet amount
    bet_amount = int(input("Enter Bet: "))
    
    # Passes the bet_amount into the place_bet func, returns a list with the amount of Chip objects the player bet
    placed_bet = place_bet(bet_amount, player.money)
    
    # Create the lists that are going to hold the player and dealer cards
    player_cards = []
    dealer_cards = []
    
    # Deals two cards from the new_deck using the deal_one function, appending them to the cards list
    for i in range(2):
        player_cards.append(new_deck.deal_one())
        dealer_cards.append(new_deck.deal_one())
    
    # Prints the two cards the player has, and the dealer's face up card
    print(f"Player cards: [{player_cards[0]}, {player_cards[1]}]")
    print(f"Dealer cards: [Hidden Card, {dealer_cards[1]}]\n")
    
    # -------------------------- PLAYER LOGIC -------------------------- #
    
    # A variable that holds the score the player has from his first 2 cards
    player_score = player_cards[0].value + player_cards[1].value
    
    # Declares bust as False, as long as the player doesn't bust or stay they can keep hitting
    bust = False
    
    print("------ PLAYER TURN ------\n")
    while not bust:
        
        print(f"Current score is: {player_score}")
        hit_stay = input("Hit (H) or Stay(S): ")
        
        # Keeps appending cards to player_cards if the player keeps typing H
        if (hit_stay == "H"):
            player_cards.append(new_deck.deal_one())
            player_score += player_cards[-1].value
            print(f"New card: {player_cards[-1]}\n")
            
            # Checks to see if the score has gone over 21, if so, breaks out of the loop
            if player_score > 21:
                for i in range(len(player_cards)):
                    if player_cards[i].rank == "Ace":
                        player_score -= 10
                        break
                if player_score > 21:
                    bust = True
                    print(f"Player bust with {player_score}")
                    break
                
        # If the player decides to stay, the loop breaks without declaring bust as True
        else:
            break

    # -------------------------- DEALER LOGIC -------------------------- #
            
    # Reveals the dealer's cards
    dealer_score = dealer_cards[0].value + dealer_cards[1].value
    print(f"\n\nDealer's cards: [{dealer_cards[0]}, {dealer_cards[1]}]")
    
    # If the player busts, the dealer doesn't play, so if bust == True, dealer wins, and the code isn't executed
    print("\n------ DEALER TURN ------\n")
    if bust == False:
        #bust = False
        while not bust:
            print(f"Dealer score is: {dealer_score}")

            # Keeps appending cards to dealer_cards if the dealer isn't at 17
            if dealer_score < 17:
                dealer_cards.append(new_deck.deal_one())
                dealer_score += dealer_cards[-1].value
                print(f"New card: {dealer_cards[-1]}\n")

                # Checks to see if the score has gone over 21, if so, breaks out of the loop
                if dealer_score > 21:
                    bust = True
                    print(f"Dealer bust with {dealer_score}\n")
                    player_wins(placed_bet, player.money)
                    break

            # If the dealer has > 17 but < 21, they can't hit anymore, so it breaks out of the loop, but doesn't bust
            else:
                break
    
    # If the player busts, he loses money and there's no need to determine a winner, the program can just end or continue
    # If the dealer busts, there's already a call to the player_wins() function
    # So determining the winner is only necessary if neither busted and we need to compare their values
    if not bust:
        determine_winner(bet_amount, player_score, dealer_score)
    
    if input("\nWant to play another round? (Y) or (N): ") != "Y":
        print(f"Player ended up with {len(player.money)} chips")
        game_on = False
        break
```
