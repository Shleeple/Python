# Global variables

```python
from random import shuffle

suits = ["Spades", "Diamonds", "Clubs", "Hearts"]
ranks = ["Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace"]
values = {"Two": 2, "Three": 3, "Four": 4, "Five": 5, "Six": 6, "Seven":7 , 
          "Eight": 8, "Nine": 9, "Ten": 10, "Jack": 10, "Queen": 10, "King": 10, "Ace": 11}
```

# Defining classes

```python
# Card
class Card():
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        return f"{self.rank} of {self.suit}"
    
# Deck
class Deck():
    def __init__(self):
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                new_card = Card(suit, rank)
                self.all_cards.append(new_card)
    
    def shuffle_deck(self):
        shuffle(self.all_cards)
    
    def deal_card(self):
        return self.all_cards.pop(0)
        
    def show_deck(self):
        for card in self.all_cards:
            print(card.__str__())

# Chip   
class Chip():
    def __init__(self):
        pass

# Player
class Player():
    def __init__(self, name):
        self.name = name
        self.money = []
        for i in range(50):
            new_chip = Chip()
            self.money.append(new_chip)
```

# Defining functions

```python
def deal_cards(deck):
    dealer_cards = []
    player_cards = []
    for i in range (2):
        dealer_cards.append(deck.deal_card())
        player_cards.append(deck.deal_card())
    print(f"Dealer Cards: \n{dealer_cards[0].__str__()}, {dealer_cards[1].__str__()} - Total: {dealer_cards[0].value + dealer_cards[1].value}\n")
    print(f"Player Cards: \n{player_cards[0].__str__()}, {player_cards[1].__str__()} - Total: {player_cards[0].value + player_cards[1].value}")

def place_bet(bet_amount, total_player_money):
    if bet_amount <= len(total_player_money):
        placed_bet = []
        for i in range(bet_amount):
            placed_bet.append(total_player_money.pop())
        return placed_bet
    elif total_player_money == 0:
        print("Out of money, you lose")
    else:
        print("Not enough money, try a different amount")

def player_wins(bet_amount, total_player_money):
    print(f"Player won {bet_amount}")
    for i in range(bet_amount*2):
        new_chip = Chip()
        total_player_money.append(new_chip)

def push(bet_amount, total_player_money):
    print("Push, player will be returned his bet")
    for i in range(bet_amount):
        new_chip = Chip()
        total_player_money.append(new_chip)
        
def determine_winner(bet_amount, player_score, dealer_score):
    if player_score > dealer_score:
        player_wins(bet_amount, player1.money)
        
    elif player_score == dealer_score:
        push(bet_amount, player1.money)
```

# Game Logic

```python
# Initial deck creation
game_deck = Deck()
game_deck.shuffle_deck()

# Player creation
player1 = Player("Juanjo")
```

```python
game_on = True

while game_on:
    
    # Time for bet
    print(f"Player has {len(player1.money)} chips")
    bet = int(input("Enter a bet amount: "))
    placed_bet = place_bet(bet, player1.money)
    
    # Cards are dealt
    dealer_cards = []
    player_cards = []
    for i in range (2):
        dealer_cards.append(game_deck.deal_card())
        player_cards.append(game_deck.deal_card())
    print(f"\nDealer Cards: \nHidden Card, {dealer_cards[1].__str__()}\n")
    print(f"Player Cards: \n{player_cards[0].__str__()}, {player_cards[1].__str__()} - Total: {player_cards[0].value + player_cards[1].value}")

    player_score = player_cards[0].value + player_cards[1].value
    dealer_score = dealer_cards[0].value + dealer_cards[1].value
    bust  = False
    
    # Player turn
    while not bust:
        player_choice = input("Do you want to hit (H) or stay (S): ")
        if player_choice.lower() == "H".lower():
            player_cards.append(game_deck.deal_card())
            player_score += player_cards[-1].value
            print(f"New Card: {player_cards[-1]}")
            print(f"New Score: {player_score}")
            
            if player_score > 21:
                bust = True
                print("Over 21, bust")
                print(f"\nDealer Cards: {dealer_cards[0].__str__()}, {dealer_cards[1].__str__()}\n")
                break
        
        elif player_choice.lower() == "S".lower():
            print("Player Stayed")
            break
        
        else:
            print("That's not H or S, try again")
            bust = False
        
    # Once player is done, it's dealers turn
    if not bust:
        print(f"\nDealer Cards: {dealer_cards[0].__str__()}, {dealer_cards[1].__str__()}\n")
        print(f"Dealer Score: {dealer_score}")
        while dealer_score < 17:
            dealer_cards.append(game_deck.deal_card())
            dealer_score += dealer_cards[-1].value
            print(f"New Card: {dealer_cards[-1]}")
            print(f"New Score: {dealer_score}")
            if dealer_score > 21:
                print("Dealer bust, player wins")
                break
    
    if player_score > dealer_score and not bust:
        player_wins(bet, player1.money)
    
    elif player_score == dealer_score:
        push(bet, player1.money)
        
    elif dealer_score > 21:
        player_wins(bet, player1.money)
    
    else:
        print("Player Loses")
    
    game_on = False
```
