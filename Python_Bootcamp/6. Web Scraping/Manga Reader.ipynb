## Select the chapter number that you wanna turn into pdf (1-205)

```python
import bs4
import requests
```

```python
base_url = "https://ww2.demonslayer.site/chapter/kimetsu-no-yaiba-chapter-{}/"
```

```python
chapter = int(input("Enter a chapter number: "))
scrape_url = base_url.format(chapter)
```

```python
scrape_url
```

```python
res = requests.get(scrape_url)
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup
```

```python
pages = []
for i in soup.select('.page-break'):
    pages.append(i)
```

```python
pages
```

```python
name: "Trial_Manga_image_{}.jpg"
from fpdf import FPDF
cnt = 1
pdf = FPDF ()
pdf.set_auto_page_break (0)
for page_data in range(len(pages)-2):
    f = open("img_{}.jpg".format(cnt), 'wb')
    
    for picture in pages[page_data].select(".wp-manga-chapter-img.img-responsive.lazyload.effect-fade"):
        img_link = picture["data-src"][7:]
        img_content = requests.get(img_link).content
        f.write(img_content)
        pdf.add_page()
        pdf.image("img_{}.jpg".format(cnt))
        cnt+=1
pdf.output("Trial_manga.pdf")
f.close()
```
